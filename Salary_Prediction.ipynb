```python
X = df.drop('salary_in_usd',axis = 1)
y = df['salary_in_usd']
```

```python
X = pd.get_dummies(X)
```

```python
df.shape
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
se = StandardScaler()
X_train = se.fit_transform(X_train)
X_test = se.transform(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,root_mean_squared_error
def evaluate(model,X_train,y_train,X_test,y_test): #mkdel - linear,knn,DT,RF
  train_pred = model.predict(X_train)
  test_pred = model.predict(X_test)

  print(f"Accuracy for your {model} is")
  print(f"training R2 score for {model} is {r2_score(y_train,train_pred)}")
  print(f"testing R2 score for {model} is {r2_score(y_test,test_pred)}")

  #MAE
  print(f"Training MAE for {model} is {mean_absolute_error(y_train,train_pred)}")
  print(f"Testing MAE for {model} is {mean_absolute_error(y_test,test_pred)}")

  #RMSE
  print(f"Training RMSE for {model} is {root_mean_squared_error(y_train,train_pred)}")
  print(f"Testing RMSE for {model} is {root_mean_squared_error(y_test,test_pred)}")
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

LinearRegression_model = LinearRegression()

# Identify rows with NaN values in X_train
nan_mask_X = np.isnan(X_train).any(axis=1)

# Identify rows with NaN values in y_train
nan_mask_y = y_train.isnull()

# Combine masks to find all rows to drop for consistency
rows_to_drop = nan_mask_X | nan_mask_y

# Create cleaned versions of X_train and y_train
X_train_cleaned = X_train[~rows_to_drop]
y_train_cleaned = y_train[~rows_to_drop]

# Fit the model with the cleaned data
LinearRegression_model.fit(X_train_cleaned, y_train_cleaned)
```

```python
evaluate(LinearRegression_model,X_train,y_train,X_test,y_test)
```

```python
y_pred = LinearRegression_model.predict(X_test)
```

```python
y_pred_train = LinearRegression_model.predict(X_train)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
r2_score(y_train,y_pred_train)
```
